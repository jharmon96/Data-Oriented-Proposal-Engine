### Initialize data set

```python
# Create dataframe
sql_tbl_name = '"2016_4"'
df = pd.read_sql_query('SELECT * FROM ' + sql_tbl_name, con=conn)
print('Shape of initial df:', df.shape)
```

```python
# Recipient
# Remove columns with information regarding the recipient
cols = [c for c in df.columns if c.lower()[:9] != 'recipient']
print(len(cols))
```

```python
df = df[cols[1:len(cols)]]
df.head()
```

```python
# Code
# Remove columns with duplicative codes
cols = [c for c in df.columns if c.lower()[-4:] != 'code']
df = df[cols[:len(cols)]]
df.shape
```

```python
# More than 60 percent missing values
cols = df.columns[df.isnull().mean()>0.6]
df = df.drop(cols, axis=1)
df.shape
```

```python
# Only one unique value
for col in df.columns:
    if len(df[col].unique()) == 1:
        df.drop(col,inplace=True,axis=1)
```
